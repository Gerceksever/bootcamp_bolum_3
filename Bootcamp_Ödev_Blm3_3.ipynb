```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

import warnings

warnings.filterwarnings("ignore")
```

```python
dagilim = np.random.normal(0.5,0.1,1000)
```

```python
plt.figure(figsize=(2,1), dpi = 150)
count, bins, ignore = plt.hist(dagilim, 50, density=True)
plt.plot(bins, 1/(0.1 * np.sqrt(2 * np.pi))*np.exp( - (bins - 0.5)**2 / (2 * 0.1**2) ),
         linewidth=2, color='y')
plt.show()
```

```python
plt.figure(figsize=(5,4), dpi = 100)
nsample = 100
res = stats.probplot(dagilim, plot=plt)
plt.title("Q-Q Çizgesi")
plt.xlabel("Teorik Değerler")
plt.ylabel("Random Dagilim Veriler")
plt.show()
```
